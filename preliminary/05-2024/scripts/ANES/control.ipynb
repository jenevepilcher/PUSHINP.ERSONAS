```python
import pandas as pd

# reading csv
anes_df = pd.read_csv("anes_timeseries_2020_csv_20220210.csv")
anes_df.shape
```

```python
## 2 dictionaries to create two tables
stance_columns = {  ##major political disagreements
            'V201200': 'leaning',
            'V201336':'abortion',
            'V202337': 'gunrights',
            'V201235': 'taxwaste',
            'V201429': 'riotsolution',
            'V201246': 'govservices',
            'V202378': 'healthservices',
            }

# renaming columns in anes_df
anes_df.rename(columns=stance_columns, inplace=True)
#creating new dataframes of selected data
stances = anes_df[list(stance_columns.values())]
##helps the storage space to delete the anes dataframe now
del anes_df
```

```python
stances.sample(5)
```

```python
stances.columns
```

```python
##cleaning stances data
['leaning', 'abortion', 'gunrights', 'taxwaste', 'riotsolution',
       'govservices', 'healthservices']
stances = stances.query('0 < leaning <= 7')
stances = stances.query('0 < abortion <= 4')
stances = stances.query('0 < gunrights <= 3')
stances = stances.query('0 < taxwaste <= 3')
stances = stances.query('0 < riotsolution <= 7')
stances = stances.query('0 < govservices <= 7')
stances = stances.query('0 < healthservices <= 3')
stances
##less rows in this data means people prefer to answer questions regarding feelings about people over their political beliefs
```
